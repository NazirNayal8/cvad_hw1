```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.figure_factory as ff
import pandas as pd
from expert_dataset import ExpertDataset
```

```python
dataset = ExpertDataset(data_root='/media/nazirnayal/DATA/datasets/Carla/expert_data/train')
```

```python
data = pd.DataFrame()
```

```python
data['speeds'] = dataset.speeds
data['commands'] = dataset.commands
data['throttle'] = dataset.actions[:, 0]
data['steer'] = dataset.actions[:, 1]
data['break'] = dataset.actions[:, 2]
```

```python
# fig = ff.create_distplot([data['throttle']], ['throttle'], bin_size=0.01)

# fig.show()
```

```python
plt.figure(figsize=(8, 5))
fig = data['speeds'].hist(log=True, bins=100)
plt.xlabel('Speed')
plt.ylabel('Log Frequency')
plt.title('Speed Log Distribution')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
fig = data['commands'].astype(np.int32).hist(log=True)
plt.xlabel('Commands')
plt.ylabel('Log Frequency')
plt.title('Commands Log Distribution')
plt.show()
```
